# Working with Palettes

The `Palette` class manages collections of colors with 22 built-in palettes.

```python
from pychromatic import Palette, Color
```

## Creating a Palette

```python
p = Palette("default")
print(p)
p.show()
```

```python
p = Palette("rainbow")
p.show()
```

```python
p = Palette("earth")
p.show()
```

## Viewing All Palettes

```python
from pychromatic.colors import color_palettes

for name in color_palettes:
    p = Palette(name)
    p.show(title=name)
```

## Detailed View

Use `minimal=False` to see line plots, pie charts, and bar charts.

```python
p = Palette("set2")
p.show(minimal=False, title="set2")
```

## Container Protocol

Palettes support `len()`, indexing, and iteration.

```python
p = Palette("default")

print(f"Number of colors: {len(p)}")
print(f"First color: {p[0]}")
print(f"By name: {p['red']}")

for i, color in enumerate(p):
    if i >= 5:
        break
    print(f"  {color.name}: {color.hex}")
```

## Switching Palettes

```python
p = Palette("default")
print(f"Before: {p}")

p.palette = "google"
print(f"After: {p}")
p.show()
```

## Random Palette

```python
p = Palette("random")
print(p)
p.show()
```

## Adding & Removing Colors

```python
p = Palette("set2")
print(f"Before: {len(p)} colors")
p.show()

p.add_color("#9c27b0", name="violet")
print(f"After add: {len(p)} colors")
p.show()
```

```python
p.remove_color("violet")
print(f"After remove: {len(p)} colors")
p.show()
```

## Brighten & Darken Palette Colors

```python
p = Palette("dark")
p.show()

p.brighten("color1", fraction=0.4, name="light_color1")
p.darken("color2", fraction=0.3, name="dark_color2")
p.show()
```

## Intermediate Colors

```python
p = Palette("set2")
p.show()

p.get_intermediate_colors("color1", "color2", num_colors=3)
p.show()
```

## Palette Colormaps

```python
import numpy as np
import matplotlib.pyplot as plt

p = Palette("rainbow")
cmap = p.get_cmap()

data = np.random.randn(20, 20)
plt.figure(figsize=(6, 4))
plt.imshow(data, cmap=cmap, aspect="auto")
plt.colorbar(label="value")
plt.title("Heatmap with rainbow palette colormap")
plt.show()
```

## Reset

```python
p = Palette("set2")
original_hex = [c.hex for c in p]

for c in p:
    c.brighten(0.3)

modified_hex = [c.hex for c in p]
print(f"Modified: {modified_hex}")

p.reset()
reset_hex = [c.hex for c in p]
print(f"Reset:    {reset_hex}")
print(f"Match original: {original_hex == reset_hex}")
```
