# Colorblind-Friendly Palettes

pychromatic ships four accessibility-first palettes.

```python
from pychromatic import Palette, okabe_ito, tableau10, tol_bright, tol_muted
from pychromatic.cutils import plot_colors
```

## Okabe-Ito

The classic 8-color palette.

```python
p = Palette("okabe_ito")
p.show()
print(f"{len(p)} colors: {[c.name for c in p]}")
```

## Tableau 10

```python
p = Palette("tableau10")
p.show()
print(f"{len(p)} colors: {[c.name for c in p]}")
```

## Tol Bright

```python
p = Palette("tol_bright")
p.show()
print(f"{len(p)} colors: {[c.name for c in p]}")
```

## Tol Muted

```python
p = Palette("tol_muted")
p.show()
print(f"{len(p)} colors: {[c.name for c in p]}")
```

## Using as Dictionaries

```python
print("Okabe-Ito:")
for name, hex_val in okabe_ito.items():
    print(f"  {name:12s} {hex_val}")

plot_colors(list(okabe_ito.values()), minimal=True)
```

```python
print("Tableau 10:")
for name, hex_val in tableau10.items():
    print(f"  {name:12s} {hex_val}")

plot_colors(list(tableau10.values()), minimal=True)
```

## Using in Plots

```python
import numpy as np
import matplotlib.pyplot as plt

p = Palette("okabe_ito")

fig, ax = plt.subplots(figsize=(8, 4))
x = np.linspace(0, 2 * np.pi, 100)

for i, color in enumerate(p):
    ax.plot(x, np.sin(x + i * 0.4), color=color.hex, linewidth=2.5, label=color.name)

ax.legend(loc="upper right", ncol=2)
ax.set_title("Okabe-Ito palette")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)

for ax, name in zip(axes, ["okabe_ito", "tableau10", "tol_bright", "tol_muted"]):
    pal = Palette(name)
    x = np.linspace(0, 2 * np.pi, 100)
    for i, color in enumerate(pal):
        ax.plot(x, np.sin(x + i * 0.4), color=color.hex, linewidth=2.5)
    ax.set_ylabel(name, fontsize=10)
    ax.set_xlim(0, 2 * np.pi)

axes[-1].set_xlabel("x")
fig.suptitle("Colorblind-Friendly Palettes", fontsize=14)
plt.tight_layout()
plt.show()
```

## Colormaps from Colorblind Palettes

```python
from pychromatic import palette_cmap

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
data = np.random.randn(15, 15)

for ax, name in zip(axes.flat, ["okabe_ito", "tableau10", "tol_bright", "tol_muted"]):
    cmap = palette_cmap(name)
    im = ax.imshow(data, cmap=cmap, aspect="auto")
    ax.set_title(name)
    plt.colorbar(im, ax=ax)

plt.suptitle("Colormaps from colorblind palettes", fontsize=14)
plt.tight_layout()
plt.show()
```
