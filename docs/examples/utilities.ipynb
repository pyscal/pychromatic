# Utility Functions

Low-level functions for color conversion, manipulation, and visualization.

```python
from pychromatic.cutils import (
    hex_to_rgb, rgb_to_hex,
    rgb_to_hls, hls_to_rgb,
    brighten, mix_colors,
    find_intermediate_colors,
    create_colormap, get_color,
    plot_colors
)
```

## Color Conversions

### Hex to/from RGB

```python
rgb = hex_to_rgb("#ff5722")
print(f"#ff5722 -> RGB: {rgb}")

hexval = rgb_to_hex([255, 87, 34])
print(f"[255, 87, 34] -> Hex: {hexval}")

hexval = rgb_to_hex([1.0, 0.0, 0.0])
print(f"[1.0, 0.0, 0.0] -> Hex: {hexval}")
```

### RGB to/from HLS

```python
hls = rgb_to_hls([255, 0, 0])
print(f"Red RGB -> HLS: {hls}")

rgb = hls_to_rgb([0.0, 0.5, 1.0])
print(f"HLS [0, 0.5, 1] -> RGB: {rgb}")
```

## Brighten & Darken

```python
base = "#1976d2"
shades = [
    brighten(base, -0.5),
    brighten(base, -0.25),
    base,
    brighten(base, 0.25),
    brighten(base, 0.5),
]
print("Dark <- -> Light")
plot_colors(shades, minimal=True)
```

```python
black_shades = [brighten("#000000", f) for f in [0.0, 0.2, 0.4, 0.6, 0.8]]
print("Black -> Light")
plot_colors(black_shades, minimal=True)
```

## Mix Colors

```python
c1 = "#d32f2f"
c2 = "#1976d2"

blends = [mix_colors(c1, c2, ratio=r) for r in [1.0, 0.75, 0.5, 0.25, 0.0]]
print("Red 100% <- -> Blue 100%")
plot_colors(blends, minimal=True)
```

## Intermediate Colors

```python
gradient = find_intermediate_colors("#d32f2f", "#388e3c", colors=8, ignore_edges=False)
print(f"Red -> Green gradient ({len(gradient)} colors):")
plot_colors(gradient, minimal=True)
```

```python
middle = find_intermediate_colors("#d32f2f", "#388e3c", colors=5, ignore_edges=True)
print(f"Intermediates only ({len(middle)} colors):")
plot_colors(middle, minimal=True)
```

## Custom Colormaps

```python
import numpy as np
import matplotlib.pyplot as plt

cmap2 = create_colormap(["#1976d2", "#ff5722"])
cmap3 = create_colormap(["#d32f2f", "#ffffff", "#1976d2"])

data = np.random.randn(12, 12)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

im1 = ax1.imshow(data, cmap=cmap2, aspect="auto")
ax1.set_title("Two-stop")
plt.colorbar(im1, ax=ax1)

im2 = ax2.imshow(data, cmap=cmap3, aspect="auto")
ax2.set_title("Diverging (3-stop)")
plt.colorbar(im2, ax=ax2)

plt.tight_layout()
plt.show()
```

## Plot Colors

Visualize any list of colors.

```python
colors = ["#d32f2f", "#1976d2", "#388e3c", "#f57c00", "#7b1fa2"]

print("Minimal view:")
plot_colors(colors, minimal=True)

print("\nFull view:")
plot_colors(colors, minimal=False)
```
