# Utility Functions

Low-level helpers in `pychromatic.cutils` for colour conversions and
visualisation.  Higher-level operations like mixing, brightening, and
colormaps are covered in the [Colors](colors.ipynb) and
[Colormaps](colormaps.ipynb) pages.

```python
from pychromatic.cutils import (
    hex_to_rgb, rgb_to_hex,
    rgb_to_hls, hls_to_rgb,
    plot_colors,
)
```

## Hex ↔ RGB

```python
rgb = hex_to_rgb("#ff5722")
print(f"#ff5722 -> RGB: {rgb}")

hexval = rgb_to_hex([255, 87, 34])
print(f"[255, 87, 34] -> Hex: {hexval}")

# Float inputs (0-1 range) are also accepted
hexval = rgb_to_hex([1.0, 0.0, 0.0])
print(f"[1.0, 0.0, 0.0] -> Hex: {hexval}")
```

## RGB ↔ HLS

```python
hls = rgb_to_hls([255, 0, 0])
print(f"Red RGB -> HLS: {hls}")

rgb = hls_to_rgb([0.0, 0.5, 1.0])
print(f"HLS [0, 0.5, 1] -> RGB: {rgb}")
```

## Visualising colour lists

`plot_colors()` renders any list of hex strings as a swatch.

```python
colors = ["#d32f2f", "#1976d2", "#388e3c", "#f57c00", "#7b1fa2"]

print("Minimal view:")
plot_colors(colors, minimal=True)

print("\nFull view:")
plot_colors(colors, minimal=False)
```
