# Colormaps & Color Sampling

pychromatic provides functions for working with matplotlib colormaps.

```python
import numpy as np
import matplotlib.pyplot as plt
from pychromatic import get_color, palette_cmap, Palette
from pychromatic.cutils import create_colormap, plot_colors
```

## Sampling from Matplotlib Colormaps

Use `get_color()` to sample a hex color at any position (0–1).

```python
viridis_colors = [get_color(v, "viridis") for v in np.linspace(0, 1, 10)]
plot_colors(viridis_colors, minimal=True)
```

```python
for cmap_name in ["viridis", "plasma", "coolwarm", "Spectral"]:
    colors = [get_color(v, cmap_name) for v in np.linspace(0, 1, 12)]
    print(cmap_name)
    plot_colors(colors, minimal=True)
```

## Palette Colormaps

Convert any palette to a matplotlib colormap with `palette_cmap()`.

```python
cmap = palette_cmap("rainbow")

sampled = [cmap(v) for v in np.linspace(0, 1, 20)]
hex_colors = [f"#{int(r*255):02x}{int(g*255):02x}{int(b*255):02x}" for r, g, b, _ in sampled]
plot_colors(hex_colors, minimal=True)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
data = np.random.randn(15, 15)

for ax, name in zip(axes.flat, ["rainbow", "set5", "earth", "accent"]):
    cmap = palette_cmap(name)
    im = ax.imshow(data, cmap=cmap, aspect="auto")
    ax.set_title(name)
    plt.colorbar(im, ax=ax)

plt.tight_layout()
plt.show()
```

## Custom Colormaps

```python
div_cmap = create_colormap(["#d32f2f", "#ffffff", "#1976d2"])

data = np.random.randn(15, 15)
plt.figure(figsize=(6, 4))
plt.imshow(data, cmap=div_cmap, aspect="auto")
plt.colorbar(label="value")
plt.title("Custom diverging colormap")
plt.show()
```

```python
multi_cmap = create_colormap(["#1a237e", "#00bcd4", "#ffeb3b", "#ff5722"])

x = np.linspace(0, 4 * np.pi, 200)
y = np.linspace(0, 4 * np.pi, 200)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

plt.figure(figsize=(6, 4))
plt.contourf(X, Y, Z, levels=30, cmap=multi_cmap)
plt.colorbar()
plt.title("Multi-stop custom colormap")
plt.show()
```

## Combining Palettes with Matplotlib

```python
np.random.seed(42)
x = np.random.randn(200)
y = np.random.randn(200)
c = np.sqrt(x**2 + y**2)

cmap = palette_cmap("set5")

plt.figure(figsize=(6, 5))
sc = plt.scatter(x, y, c=c, cmap=cmap, s=50, edgecolors="none")
plt.colorbar(sc, label="distance from origin")
plt.title("Scatter plot with set5 palette colormap")
plt.show()
```
