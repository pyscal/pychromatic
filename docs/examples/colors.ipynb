# Working with Colors

The `Color` class is the fundamental building block of pychromatic.

```python
from pychromatic import Color
```

## Creating Colors

Create a color from a hex code.

```python
red = Color("#d32f2f", name="red")
blue = Color("#1976d2", name="blue")
green = Color("#388e3c", name="green")
print(red)
print(blue)
print(green)
```

## Displaying Colors

Use `.show()` to render a color swatch.

```python
red.show()
```

```python
blue.show()
```

## Color Attributes

```python
c = Color("#ff5722", name="deep_orange")
print(f"Hex: {c.hex}")
print(f"RGB: {c.rgb}")
print(f"HLS: {c.hls}")
```

Unpack RGB values since `Color` supports iteration:

```python
r, g, b = c
print(f"r={r}, g={g}, b={b}")
```

## Brighten & Darken

```python
import matplotlib.pyplot as plt
from pychromatic.cutils import plot_colors, brighten

base = "#1976d2"
shades = [
    brighten(base, -0.4),
    brighten(base, -0.2),
    base,
    brighten(base, 0.2),
    brighten(base, 0.4),
]
plot_colors(shades, minimal=True)
```

Using `.brighten()` and `.darken()` on a Color object:

```python
c = Color("#1976d2", name="blue")
c.brighten(0.3)
c.show()
print(f"Brightened: {c.hex}")
```

```python
c.reset()
c.darken(0.3)
c.show()
print(f"Darkened: {c.hex}")
```

## Multiplying Colors

Use `*` operator — values < 1 brighten, values > 1 darken.

```python
c = Color("#d32f2f", name="red")
lighter = c * 0.3
darker = c * 1.3

plot_colors([darker.hex, c.hex, lighter.hex], minimal=True)
```

## Mixing Colors

```python
from pychromatic.cutils import mix_colors

c1 = "#d32f2f"
c2 = "#1976d2"

blends = [mix_colors(c1, c2, ratio=r) for r in [1.0, 0.75, 0.5, 0.25, 0.0]]
plot_colors(blends, minimal=True)
```

## Intermediate Colors

```python
from pychromatic.cutils import find_intermediate_colors

gradient = find_intermediate_colors("#d32f2f", "#1976d2", colors=8, ignore_edges=False)
plot_colors(gradient, minimal=True)
```

## Color Equality

Colors are compared by hex value (case-insensitive) and are hashable.

```python
c1 = Color("#ff0000")
c2 = Color("#ff0000", name="red")
c3 = Color("#0000ff")

print(f"c1 == c2: {c1 == c2}")
print(f"c1 == c3: {c1 == c3}")
print(f"Unique colors in set: {len({c1, c2, c3})}")
```

## Input Validation

```python
try:
    Color("not-a-color")
except ValueError as e:
    print(f"Caught: {e}")
```
