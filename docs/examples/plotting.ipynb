# Plotting Utilities

`Multiplot`, `BrokenAxes`, and the `chromatify` decorator for
publication-quality figures.

```python
import numpy as np
import matplotlib.pyplot as plt
from pychromatic import Multiplot, BrokenAxes, chromatify, Palette
```

## Multiplot

Create flexible subplot layouts. The constructor accepts `rows`, `columns`,
and `width` (in points). Axes are pre-created and accessed via `m[row, col]`.

```python
m = Multiplot(rows=2, columns=2, width=700, ratio=0.8)

for r in range(2):
    for c in range(2):
        ax = m[r, c]
        x = np.linspace(0, 2*np.pi, 100)
        ax.plot(x, np.sin(x + r + c), linewidth=2)
        ax.set_title(f"Subplot ({r},{c})")

plt.tight_layout()
plt.show()
```

### Custom Sizing with width_ratios

```python
m = Multiplot(rows=1, columns=2, width=800, ratio=0.4,
              width_ratios=[2, 1])

ax1 = m[0, 0]
x = np.linspace(0, 4*np.pi, 200)
ax1.plot(x, np.sin(x), linewidth=2)
ax1.set_title("Wide plot")

ax2 = m[0, 1]
ax2.bar([1, 2, 3], [4, 7, 2], color=["#d32f2f", "#1976d2", "#388e3c"])
ax2.set_title("Narrow plot")

plt.tight_layout()
plt.show()
```

### Adding Insets

Use `add_inset(index)` to place a small axes inside an existing subplot.

```python
m = Multiplot(rows=1, columns=1, width=600)

ax = m[0, 0]
x = np.linspace(0, 10, 300)
ax.plot(x, np.sin(x) * np.exp(-x/5), linewidth=2, color="#1976d2")
ax.set_title("Main plot")

m.add_inset((0, 0), width="30%", height="30%", loc=1)
inset = m[0, 0, 0]
inset.plot(x[:50], np.sin(x[:50]), color="#d32f2f", linewidth=1.5)
inset.set_title("Zoomed in", fontsize=9)

plt.tight_layout()
plt.show()
```

### Styling Axes

`style_axes(index)` provides a convenient way to customise labels, titles, spines, and ticks.

```python
m = Multiplot(rows=1, columns=2, width=800, ratio=0.4)

ax1 = m[0, 0]
ax1.plot([1, 2, 3], [1, 4, 2], linewidth=2)
ax1.set_title("Default style")

ax2 = m[0, 1]
ax2.plot([1, 2, 3], [1, 4, 2], linewidth=2)
m.style_axes(
    (0, 1),
    title="Custom style",
    xlabel="x", ylabel="y",
    labelsize=12,
    hide_spines=["top", "right"],
)

plt.tight_layout()
plt.show()
```

## chromatify Decorator

Wrap any plotting function with `@chromatify` to auto-apply pychromatic's colour cycle.

```python
fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(0, 2 * np.pi, 100)
for i in range(6):
    ax.plot(x, np.sin(x + i * 0.5), linewidth=2, label=f"Series {i+1}")

ax.legend()
ax.set_title("Default matplotlib colors")
plt.tight_layout()
plt.show()
```

```python
@chromatify
def my_plot():
    fig, ax = plt.subplots(figsize=(8, 4))
    x = np.linspace(0, 2 * np.pi, 100)
    for i in range(6):
        ax.plot(x, np.sin(x + i * 0.5), linewidth=2, label=f"Series {i+1}")
    ax.legend()
    ax.set_title("With @chromatify")
    plt.tight_layout()
    return ax

my_plot()
plt.show()
```

## BrokenAxes

Standalone broken-axis plots for discontinuous data ranges.

### Broken X-Axis

Pass `x` as four values `[lo1, hi1, lo2, hi2]` and `y` as two values `[lo, hi]`.

```python
ba = BrokenAxes(x=[0, 5, 15, 20], y=[-1.2, 1.2], width=600)

x1 = np.linspace(0, 5, 50)
x2 = np.linspace(15, 20, 50)

# axes[1] = left panel, axes[2] = right panel
ba.axes[1].plot(x1, np.sin(x1), color="#1976d2", linewidth=2)
ba.axes[2].plot(x2, np.sin(x2), color="#1976d2", linewidth=2)

# axes[0] is the hidden master axes
ba.axes[0].set_title("Broken X-Axis", fontsize=14, pad=15)

plt.show()
```

### Broken Y-Axis

Pass `y` as four values `[lo1, hi1, lo2, hi2]` and `x` as two values `[lo, hi]`.

```python
x_vals = np.arange(10)
y_vals = np.array([2, 3, 2.5, 3.5, 2, 3, 50, 55, 52, 48])

ba = BrokenAxes(x=[0, 9], y=[0, 5, 45, 60], width=600)

ba.axes[1].plot(x_vals, y_vals, "o-", color="#d32f2f", linewidth=2)
ba.axes[2].plot(x_vals, y_vals, "o-", color="#d32f2f", linewidth=2)

ba.axes[0].set_title("Broken Y-Axis", fontsize=14, pad=15)

plt.show()
```

### Broken Axes inside Multiplot

`Multiplot.add_brokenaxes()` embeds broken axes in a subplot cell.

```python
m = Multiplot(rows=1, columns=2, width=800, ratio=0.5)

# regular plot on the left
ax = m[0, 0]
x = np.linspace(0, 2*np.pi, 100)
ax.plot(x, np.sin(x), linewidth=2, color="#1976d2")
ax.set_title("Normal")

# broken x-axis on the right
m.add_brokenaxes((0, 1), x=[0, 5, 15, 20], y=[-1.2, 1.2])
bax = m[0, 1, 0]  # first sub-axis
bbx = m[0, 1, 1]  # second sub-axis

x1 = np.linspace(0, 5, 50)
x2 = np.linspace(15, 20, 50)
bax.plot(x1, np.sin(x1), color="#d32f2f", linewidth=2)
bbx.plot(x2, np.sin(x2), color="#d32f2f", linewidth=2)

plt.show()
```
